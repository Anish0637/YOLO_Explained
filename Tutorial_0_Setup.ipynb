### Download the pretrained yolo v2 model 
The inventor of Yolo has a website listing all variations of Yolo V2 model. [Check it out](https://pjreddie.com/darknet/yolov2/)!

We are going to use the **YOLOv2	VOC 2007+2012** model, which is trained on the Pascal VOC dataset. Download the model and put it in the **root directory of the repo** (Otherwise you might need to change the path variables in the tutorials)
1. Yolo V2: https://pjreddie.com/media/files/yolov2-voc.weights
2. Pretrained Darknet 19 448 model: http://pjreddie.com/media/files/darknet19_448.weights (use for transfer learning)

### Convert the yolo v2 weights to tensorlfow .h5 file 

```python
CONFIG_PATH = 'Metadata/PascalVoc12/yolov2Voc.cfg' # a file that describes the yolo architecture. Downloaded from the yolo website
WEIGHT_PATH = 'yolov2-voc.weights' # the file you just downloaded
from YoloBackbone.modelBuilder import YOLO2
builder = YOLO2(configPath=CONFIG_PATH, weightPath=WEIGHT_PATH, inputShape=None)
model, anchors, classNum = builder.buildModel() # now we have a tensorflow model, don't worry about the other two for now
```

```python
#save the tensorflow model in the root directory
model.save('yolov2_voc.h5')
```
